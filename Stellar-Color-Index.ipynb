```python
import csv
with open('data/hygfull.csv') as f:
    stars = list(csv.DictReader(f))

print 'The database holds', len(stars), 'stars'
print 'An example star:', stars[123]
```

```python
import spectral_classification
color_chart = spectral_classification.build_color_chart('starcolors.txt')

print 'The RGB color for class G2(V) is:\n', color_chart['G2(V)']
```

```python
useful_stars = [
    star for star in stars
    if star['Spectrum'] in color_chart and
       star['ColorIndex'] != ''
    ]

print 'There are', len(useful_stars), '/', len(stars), 'useful stars'
print 'Which is', 100.0 * float(len(useful_stars)) / len(stars), '%'
```

```python
color_index = np.array(
    [ float(star['ColorIndex']) for star in useful_stars ])
spectral_code = np.array(
    [ star['Spectrum'] for star in useful_stars ])

print 'First few spectral codes:', spectral_code[:5]
print 'First few color indexes:', color_index[:5]
```

```python
rgb_color = np.array(
    [ color_chart[code] for code in spectral_code ])

print 'First few RGB colors:\n', rgb_color[:5]
```

```python
red = rgb_color[:,0]
green = rgb_color[:,1]
blue = rgb_color[:,2]

big = (8.0, 16.0)
subplot = pl.figure(figsize=big).add_subplot

subplot(311, aspect=4).scatter(color_index, red, s=1.0)
subplot(312, aspect=4).scatter(color_index, green, s=1.0)
subplot(313, aspect=4).scatter(color_index, blue, s=1.0)
```

```python
result = stats.linregress(color_index, red)
slope, intercept, r_value, p_value, std_err = result

print slope, intercept, r_value, p_value, std_err

plot = pl.figure(figsize=big).add_subplot(311)
plot.scatter(color_index, red, s=1.0)
x0, x1, y0, y1 = plot.axis()
plot.plot([x0, x1], [intercept + x0 * slope,
                     intercept + x1 * slope])
```

```python
def run_regression(color):
    result = stats.linregress(color_index[color < 1.0],
                              color[color < 1.0])
    slope, intercept, r_value, p_value, std_err = result

    print slope, intercept, r_value, p_value, std_err

    plot = pl.figure(figsize=big).add_subplot(311)
    plot.scatter(color_index, color, s=1.0)
    x0, x1, y0, y1 = plot.axis()
    plot.plot([x0, x1], [intercept + x0 * slope,
                         intercept + x1 * slope])

run_regression(red)
```
